```python
#C R U D
# create a collection in Chromadb
# collection is a collection of docs
import chromadb
client = chromadb.Client()
fruits = client.create_collection("fruit")
```

```python
## creating a empty container ##
print(type(fruits))
```

```python
## now filling documents inside the container named fruits ##
fruits.add(
    ids=["1", "2", "3", "4", "5", "6", "7"],
    documents=["apple",
        "banana",
        "pineapple",
        "mango",
        "dragonfruit",
        "passionfruit",
        "raspberry"],
    metadatas=[
        { "color": "red", "weight": 180 },
        { "color": "yellow", "weight": 120 },
        { "color": "brown", "weight": 900 },
        { "color": "yellow", "weight": 200 },
        { "color": "pink", "weight": 600 },
        { "color": "purple", "weight": 18 },
        { "color": "red", "weight": 4 },   
    ]
)
```

```python
# Read the data
all_data = fruits.get()
print(all_data)
```

```python
##pprint- pretty print- to show o/p nicer with key value pairs ####
from pprint import pprint
pprint(all_data)
```

```python
## now exclusively mention to include embedding- as by default collection wont shown as above##

all_data_with_embeddings = fruits.get(include=["embeddings", "documents", "metadatas"])
```

```python
print(all_data_with_embeddings)
```

```python
pprint(all_data_with_embeddings)
```

```python
# Now that data is stored with embeddings, lets query and so semantic search!
# Semantic Search
tropical_fruits = fruits.query(query_texts=["Tropical"])
```

```python
pprint(tropical_fruits)
```

```python
# Print required data
print(tropical_fruits["documents"])
```

```python
# Metadata Filtering
tropical_yellow = fruits.query(query_texts=["Tropical"], 
                                  where={"$and":[
                                      {"color": "yellow"},
                                      {"weight": {"$gt": 150}}
                                  ]
                                  })
```

```python
print(tropical_yellow["documents"])
```

```python
# Full Text Search
contains_fruit = fruits.get(
    where_document={"$contains": "fruit"}
)
print(contains_fruit["documents"])
```

```python
import openai
from chromadb.utils.embedding_functions import OpenAIEmbeddingFunction
from dotenv import load_dotenv
```

```python
load_dotenv()
embedding_func = OpenAIEmbeddingFunction(
    model_name="text-embedding-3-small",  # or "text-embedding-3-large"
    api_key=""
)
```

```python
# Types of Storage
# Ephemeral Client - What we did so far, NOT persisted
# Persistent Client - Saved to Disk
# Cloud Client - Saved to Cloud
```

```python
# We will recreate the Fruits database with persistent client
persistent_directory = "db"
persistent_client = chromadb.PersistentClient(path=persistent_directory)
```

```python
print(client.list_collections())
```

```python
persistent_collection = persistent_client.create_collection("fruit", embedding_function=embedding_func)
```

```python
persistent_collection.add(
    ids=["1", "2", "3", "4", "5", "6", "7"],
    documents=["apple",
        "banana",
        "pineapple",
        "mango",
        "dragonfruit",
        "passionfruit",
        "raspberry"],
    metadatas=[
        { "color": "red", "weight": 180 },
        { "color": "yellow", "weight": 120 },
        { "color": "brown", "weight": 900 },
        { "color": "yellow", "weight": 200 },
        { "color": "pink", "weight": 600 },
        { "color": "purple", "weight": 18 },
        { "color": "red", "weight": 4 },        
    ]
)
```
